```python
# 기본 그래프 그리기 
x = [1,3,5,7,9]
y = [100,200,300,400,500]
plt.plot(x, y)
```

```python
# 코사인 함수 그리기
x = np.linspace(-np.pi, np.pi, 128) # 연속적인 값을 갖는 배열 (-pi부터 pi까지 128개의 원소로 만들라)
y = np.cos(x)
plt.plot(y)
```
